# 1. importing necessasry libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# 2. web page request from the website as an user

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?pages=1',headers=headers).text
```

# 3. getting websites code with indentation

```python
soup=BeautifulSoup(webpage,'lxml')
##print(soup.prettify())
```

# 4. scrapping data from the first page

```python
soup.find_all('h1')[0].text
```

```python
company = soup.find_all('div',class_='company-content-wrapper')
```

```python
len(company)
```

```python
name=[]
rating=[]
review=[]
company_type=[]
hq=[]
old=[]
for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p',class_='rating').text.strip())
    review.append(i.find('a',class_="review-count").text.strip())
    company_type.append(i.find_all('p',class_="infoEntity")[1].text.strip())
    hq.append(i.find_all('p',class_="infoEntity")[2].text.strip())
    old.append(i.find_all('p',class_="infoEntity")[3].text.strip())
```

```python
d = {'name':name , "rating":rating , "review":review , "company_type":company_type , "head_quarter":hq , "old":old}
```

# 5. scrapping data from multiple pages

```python
final_df = pd.DataFrame()

for j in range(1,100):
    url = 'https://www.ambitionbox.com/list-of-companies?pages{}'.format(j)
    
    header = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    web_page = requests.get(url,headers=headers).text
    
    soup = BeautifulSoup(web_page,'lxml')
    
    company = soup.find_all('div',class_='company-content-wrapper')
    
    name=[]
    rating=[]
    review=[]
    company_type=[]
    hq=[]
    old=[]
    employee=[]
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p',class_='rating').text.strip())
        review.append(i.find('a',class_="review-count").text.strip())
        company_type.append(i.find_all('p',class_="infoEntity")[0].text.strip())
        hq.append(i.find_all('p',class_="infoEntity")[1].text.strip())
        old.append(i.find_all('p',class_="infoEntity")[2].text.strip())
        employee.append(i.find_all('p',class_="infoEntity")[3].text.strip())
        
    d = {'name':name , "rating":rating , "review":review , "company_type":company_type , "head_quarter":hq , "old":old ,"employee":employee}
    
    df=pd.DataFrame(d)
    
    final_df = final_df.append(df,ignore_index=True)
```
